```python
import os

dosya_adi="gorevler.txt"

def gorevleri_yukle():

  if not os.path.exists(dosya_adi):
       return []
  try:
        with open(dosya_adi,"r", encoding="utf-8")as dosya:
         gorevler=[satir.strip() for satir in dosya.readlines()]
        print("Görevler dosyadan başarıyla yüklendi.")
        return gorevler
  except Exception as e:
          print(f"Görevler yüklenirken bir hata oluştu:{e}")
          return[]

def gorevleri_kaydet(gorevler):

 try:
     with open(dosya_adi,"w", encoding="utf-8") as dosya:
        for gorev in gorevler:
          dosya.write(gorev + "\n")
     return True
 except Exception as e:
          print(f"Görevler kaydedilirken bir hata oluştu: {e}")
          return False

def gorevleri_listele(gorevler):

  print("\n----GÖREV LİSTESİ ---")
  if not gorevler:
    print("Görev listeniz boş.")
  else:
      for i, gorev in enumerate(gorevler,1):
         print(f"{i}.{gorev}")
  print("--------------------------\n")

def yeni_gorev_ekle(gorevler):

  yeni_gorev=input("eklemek istediğiniz görevi girin:").strip()
  if yeni_gorev:
    gorevler.append(yeni_gorev)
    if gorevleri_kaydet(gorevler):
          print(f"'{yeni_gorev}'görev başarıyla eklendi.")
  else:
        print("HATA:BOŞ GÖREV EKLEYEMEZSİNİZ")

def gorev_duzenle(gorevler):

  gorevleri_listele(gorevler)
  if not gorevler:
    return

  try:
      gorev_no = int(input("düzenlemek istediğiniz görevin numarasını giriniz:"))
      if 1 <= gorev_no <=len(gorevler):
        yeni_metin=input(f"{gorev_no}. gorev için yeni metin giriniz:").strip()
        if yeni_metin:
          eski_gorev =gorevler[gorev_no - 1]
          gorevler[gorev_no - 1]=yeni_metin
          if gorevleri_kaydet(gorevler):
            print(f"'{eski_gorev}' görevi '{yeni_metin}' olarak güncellendi.")
          else:
              print("hata: boş görev metni giremezsiniz.")
        else:
                print("hata: geçersiz görev numarası girdiniz.")
  except ValueError:
                  print("Hata:Lütfen sayısal bir değer giriniz.")
def gorev_sil(gorevler):
  gorevleri_listele(gorevler)
  if not gorevler:
      return

  try:
      gorev_no=int(input("silmek istediğiniz görevin numarasını giriniz: "))
      if 1 <= gorev_no <= len(gorevler):
          silinen_gorev = gorevler.pop(gorev_no - 1)
          if gorevleri_kaydet(gorevler):
             print(f"'{silinen_gorev}' görevi başarıyla silindi.")

      else:
          print("Hata: Geçersiz görev numarası girdiniz.")
  except ValueError:
        print("Hata:Lütfen sayısal bir değer giriniz.")

def ana_menu():

  gorevler = gorevleri_yukle()

  while True:
    print("\n********* GÖREV YÖNETİM UYGULAMASI *********")
    print("1.GÖREVLERİ LİSTELE")
    print("2. YENİ GÖREV EKLE")
    print("3. GÖREV DÜZENLE")
    print("4. GÖREV SİL")
    print("5. ÇIKIŞ")
    print("*********************************************")

    secim=input("Lütfen bir seçenek giriniz (1-5):")

    if secim=='1':
      gorevleri_listele(gorevler)
    elif secim=='2':
        yeni_gorev_ekle(gorevler)
    elif secim=='3':
        gorev_duzenle(gorevler)
    elif secim=='4':
        gorev_sil(gorevler)
    elif secim=='5':
          print("Uygulamadan çıkılıyor Hoşça Kalın!")
          break
    else:
          print("Geçersiz seçim! Lütfen 1 ile 5 arasında bir sayı giriniz .")

if __name__=="__main__":
  ana_menu()
```
